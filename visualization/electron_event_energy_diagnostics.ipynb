```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pickle
```

```python
with open('/home/lhertel/projects/physicsprojects/nova/simple/predictions/electron_event_energy_scaled_0_1800.pkl', 'r') as f:
    results = pickle.load(f)
```

```python
fig, axes = plt.subplots()
hist = axes.hist(results['resolution'], bins=200, normed=1, range=(0,2), histtype='step')
```

```python
fig, axes = plt.subplots()
hist = axes.scatter(results['resolution']-1, results['y'], marker='.', s=0.1)
axes.set_xlabel(r'$(E_{pred} - E_{true})/E_{true}$', fontsize=15)
axes.set_ylabel(r'$E_{true}$', fontsize=15)
axes.set_title(r'$\nu_e$', fontsize=15)
axes.set_ylim([0, 20])
axes.set_xlim([-1.5, 1.5])
fig.tight_layout()
fig.savefig('figures/nue_diagnostic_1.png')
```

```python
fig, axes = plt.subplots()
hist = axes.scatter(results['y'], results['yhat'], marker='.', s=0.1)
axes.set_xlabel(r'$E_{true}$', fontsize=15)
axes.set_ylabel(r'$E_{pred}$', fontsize=15)
axes.set_title(r'$\nu_e$', fontsize=15)
axes.set_ylim([0, 20])
axes.set_xlim([0, 20])
fig.tight_layout()
fig.savefig('figures/nue_diagnostic_2.png')
```

```python
fig = plt.figure(figsize=(15, 15))
axes = fig.add_subplot(131)
axes.set_xlabel(r'$(E_{pred} - E_{true})/E_{true}$', fontsize=15)
axes.set_ylabel(r'$E_{true}$', fontsize=15)
axes.set_title(r'$\nu_e$', fontsize=15)
H, xedges, yedges = np.histogram2d(x=results['resolution'][:, 0]-1, y=results['y'][:, 0], range=[[-1.5, 1.5], [0, 20]], bins=300)
H = H.T
plt.imshow(np.log(0.25+H), interpolation='nearest', origin='low')
axes.set_xticklabels([1, 2, 3, 4, 5, 6, 7])
fig.tight_layout()
# fig.savefig('figures/nue_diagnostic_heatmap.png')
```
